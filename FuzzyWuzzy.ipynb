```python
import pandas as pd
import numpy as np

from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
df1 = pd.read_csv('df1.csv')
df2 = pd.read_csv('df2.csv')
```

```python
def fuzz_match(col_name1, col_name2, cutoff):

    list1 = []
    for i in col_name1:
        match_res1 = process.extractOne(i, col_name2, scorer = fuzz.partial_ratio, score_cutoff=cutoff)
        list1.append(match_res1)
        
    list2 = []
    for i in col_name1:
        match_res2 = process.extractOne(i, col_name2, scorer = fuzz.ratio, score_cutoff=cutoff)
        list2.append(match_res2)
    
    list5 = []
    list6 = []
    list7 = []
    list8 = []

    for i in list1:
        if i==None:
            list5.append(None)
            list6.append(None)
        else:
            list5.append(i[0])
            list6.append(i[1])

    for i in list2:
        if i==None:
            list7.append(None)
            list8.append(None)
        else:
            list7.append(i[0])
            list8.append(i[1])
            
    tempp = pd.DataFrame()
    tempp['origin'] = col_name1
    tempp['partial_ratio_match'] = pd.Series(list5)
    tempp['score1'] = pd.Series(list6)
    tempp['ratio_match'] = pd.Series(list7)
    tempp['score2'] = pd.Series(list8)
    return tempp
```

```python
a = fuzz_match(col_name1 = df1['name'], col_name2 = df2['NAME'], cutoff=70)
a
```
